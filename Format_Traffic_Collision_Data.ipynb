## Libraries

```python
import geopandas as gpd
import pandas as pd
```

## Convert data to Pandas DF and Save as Pickle

```python
geo_df = gpd.read_file('Traffic_Collisions.geojson',driver='GeoJSON')
```

```python
df= pd.DataFrame(geo_df)
```

```python
df['lat'] = df.geometry.apply(lambda p: p.y)
df['lon'] = df.geometry.apply(lambda p: p.x)
```

```python
# Convert numeric fields to numeric type
numeric_fields = ['NoInjured', 'NoKilled', 'PtyAtFault', 'PartyAge1', 'PartyAge2']
for field in numeric_fields:
    df[field] = pd.to_numeric(df[field])
    df[field] = df[field].fillna(0) # empty cells become NaNs. Replace with 0s

# Convert all numeric fields to integer
df = df.astype({
    'NoInjured': int,
    'NoKilled': int,
    'PtyAtFault': int,
    'PartyAge1': int,
    'PartyAge2': int
})
    
# Convert date to datetime object
# Time isn't always provided, so can't rely on for datetime conversion. Date is sufficient
df['datetime'] = pd.to_datetime(df['Date'], format='%m/%d/%Y')
```

```python
# save traffic collision data as a pickle object
df.to_pickle('Traffic_Collisions.pkl')
```
