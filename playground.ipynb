```python
import geopandas as gpd 
import fiona 
import numpy as np
import pandas as pd 
```

```python
fiona.drvsupport.supported_drivers['LIBKML'] = 'rw'
```

```python
geo_df = gpd.read_file('Traffic_Collisions.kml',driver='LIBKML')
```

```python
df= pd.DataFrame(geo_df)
```

```python
df['lat'] = df.geometry.apply(lambda p: p.y)
df['lon'] = df.geometry.apply(lambda p: p.x)
```

```python
import plotly.express as px
```

```python
fig = px.density_mapbox(df, lat='lat', lon='lon', z='NoInjured', radius=4,
                        center=dict(lat=34.15127, lon=-118.1369), zoom=14,
                        mapbox_style="stamen-terrain", height=1000)
fig.show()
```
